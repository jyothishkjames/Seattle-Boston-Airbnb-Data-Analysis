```python
#Compare rates between the two cities over time.
#-----------------------------------------------

#filling missing values for the field 'price' in dataframe 'seattle_calander_df' and 'boston_calander_df'
seattle_calander_df['price'].fillna(method='ffill', inplace=True)
boston_calander_df['price'].fillna(method='bfill', inplace=True)
boston_calander_df['price'].fillna(method='ffill', inplace=True)
```

```python
#check if there are any NaN values in the column 'price' in dataframe 'seattle_calander_df' and 'boston_calander_df'
seattle_calander_df[seattle_calander_df['price'].isna() == True]
boston_calander_df[boston_calander_df['price'].isna() == True]
```

```python
#adding columns for month and year
func_month = lambda x: x.split('-')[1] 
func_year = lambda x: x.split('-')[0] 

seattle_calander_df['month'] = seattle_calander_df['date'].apply(func_month)
seattle_calander_df['year'] = seattle_calander_df['date'].apply(func_year)

boston_calander_df['month'] = boston_calander_df['date'].apply(func_month)
boston_calander_df['year'] = boston_calander_df['date'].apply(func_year)
```

```python
boston_calander_df.head()
```

```python
#Removing '$' symbol from price
seattle_calander_df['price'] = seattle_calander_df['price'].str.replace('$', '')
seattle_calander_df['price'] = seattle_calander_df['price'].str.replace(',', '')

boston_calander_df['price'] = boston_calander_df['price'].str.replace('$', '')
boston_calander_df['price'] = boston_calander_df['price'].str.replace(',', '')
boston_calander_df.head()
```

```python
#Selecting only the year 2016
seattle_calander_df = seattle_calander_df[seattle_calander_df['year'] == '2016']
boston_calander_df = boston_calander_df[boston_calander_df['year'] == '2017']
```

```python
#convert 'price' and 'month' columns in dataframe 'seattle_calander_df' and 'boston_calander_df' to float type  
seattle_calander_df["price"] = pd.to_numeric(seattle_calander_df["price"])
seattle_calander_df["month"] = pd.to_numeric(seattle_calander_df["month"])

boston_calander_df["price"] = pd.to_numeric(boston_calander_df["price"])
boston_calander_df["month"] = pd.to_numeric(boston_calander_df["month"])
```

```python
boston_calander_df.head()
```

```python
#Line plot for the average price for a month over a year for Seattle
seattle_calander_df.groupby('month')['price'].mean().plot(kind='line', color='blue')
```

```python
#Line plot for the average price for a month over a year for Boston
boston_calander_df.groupby('month')['price'].mean().plot(kind='line', color='blue')
```
